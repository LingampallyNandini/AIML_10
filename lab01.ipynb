##3.Scatterplot

```python
people = ['Ann','Brandon','chen']
age = [21,12,32]
weight = [55,35,77]
height = [160,135,170]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(age,height)
plt.title("Relationship between age and height")
plt.ylabel("age in years")
plt.xlabel("Height in meters")
plt.show()
```

```python
plt.scatter(height,weight)
plt.title("Relationship height age and weight")
plt.ylabel("height in meters")
plt.xlabel("Weight in kgs")
plt.show()
```

##4.Bar Chart


```python
plt.figure(figsize=(3,4))
plt.title("Peoples weight in Kgs",fontsize=16,fontstyle='italic')
#Main plot function 'bar'
plt.bar(x=people,height=weight,width=0.6,color='orange',edgecolor='k',alpha=0.6)
plt.xlabel("people",fontsize=15)
plt.xticks(fontsize=14,rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("Weight (in kgs)",fontsize=15)
plt.show()
```

##5.Histogram

```python
import numpy as np
plt.figure(figsize=(5,4))
plt.hist(weight,color='orange',edgecolor='k',alpha=0.75,bins=5)
plt.title("Histogram of patient weight",fontsize=18)
plt.xlabel("Weight (in kgs)",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
